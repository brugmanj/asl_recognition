## Functions to be used throughout the project

```python
# Reads the image as an array
image_read = lambda file: cv2.imread(file, 0)

```

## Loop reading in all images for one letter, and organizing into a dataframe

```python
# filepath to the A pictures
a_path = '../data/asl_alphabet_train/A/'
files = os.listdir(a_path)

# array to contain all images
a_df = []

# Looping through each file in folder
# Scaling the image between 0 and 1
for i in files:
    path = a_path + i
    image = image_read(path)
    image = image/255
    a_df.append(image)
```

## Calculating the average shape of the sign

```python
sums = np.zeros((200,200))
add = lambda x, y: np.add(x, y)

for i in range(3000):
    sums = add(sums, a_df[i])

sums /= 3000
```
